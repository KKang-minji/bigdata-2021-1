```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
plt.rcParams['font.family'] = 'Gulim'
```

```python
#파일 불러오기
data_file = easygui.fileopenbox()
data_file
```

```python
#파일읽기
dt= pd.read_csv(data_file, encoding='cp949')   # encoding ='utf8'
dt.head(10)
```

```python
dt.shape
```

2.모든 변수의 기술통계량을 구하시오 -describe

```python
dt.describe(include='all')
```

3.도착지방향총교통량이 0보다 큰 자료를 df_0에 저장하고, 관측값의 수와 변수의 수를 구하시오 -loc

```python
df_0=dt.loc[dt['도착지방향총교통량']>0]
df_0.shape   #40539개
```

4. 3번에서 저장한 df_0자료에 대하여, 집계일자에 대한 막대그래프를 그리시오

```python
tb = pd.crosstab(df_0['집계일자'], 'Freq',colnames=[''])
tb
```

```python
ax =tb.plot.bar(rot=0)
```

5. df_0자료에서 8월 2일 자료만 추출하여 df_8_2에 저장하고, 관측값의 수와 변수의 수를 구하시오 - dt.loc[]


```python
df_8_2 = df_0.loc[df_0['집계일자']==20180802]
df_8_2.shape
```

6. df_8_2자료에서 출발 업소명이 안성인 자료를 추출하여 df_as에 저장하고 관측값의 수와 변수의 수를 구하시오

```python
df_as =df_8_2.loc[df_8_2['출발영업소명']=='안성']
df_as.shape
```

7. df_as자료에서 도착지방향총교통량에 대한 기술통계량을 구하시오 -df.describe

```python
df_as.describe()
```

8. df_as 자료에서 도착지방향총교통량이 20보다 큰 자료를추출하여 df_as_20에 저장하고 관측값의 수와변수의 수를 구하시오

```python
df_as_20 =df_as.loc[df_as['도착지방향총교통량']>20]
df_as_20.shape
```

9.df_as_20 자료에서 도착지방향총교통량에 대한 히스토그램을 그리시오.

```python
ax =df_as_20['도착지방향총교통량'].plot.hist()
```

10.df_as_20 자료에서 도착지방향총교통량에 대한 박스플랏을 그리시오.

```python
ax =df_as_20['도착지방향총교통량'].plot.box()
```

11. df_as_20 자료의 도착지방향 1종교통량과 도착지방향2종교통량에 대한 산점도를 그리시오

```python
ax = df_as_20.plot.scatter(x = '도착지방향1종교통량', y='도착지방향2종교통량',c='darkblue')
```
