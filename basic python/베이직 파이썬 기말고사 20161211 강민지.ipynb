1."emergency_message_real.csv" 파일은 재난문자 발송에 대한 자료이다. 주어진 파일을 df에 저장하
고 메시지(`msg`) 내용에 "확진" 단어의 수를 세어 df['cnt']에 저장하시오.

```python
data_file = easygui.fileopenbox()
data_file
df= pd.read_csv(data_file, encoding="cp949")  
df.head()
```

```python
df['cnt']= df['msg'].str.count('확진')
df['cnt']
df.head()
```

2.df의 `cnt`에 대한 원그래프를 그리시오.

```python
tb = pd.crosstab(df['cnt'], 'Freq', colnames=[''])
tb
ax = tb.plot.pie(y='Freq', legend =False, figsize = (6,6))
```

3. df의 메시지(`msg`) 내용의 처음 대괄호로 묶인 발송처를 추출하여 df['msg_send']에 저장하시오.

```python
df['msg_send']= df['msg'].str.split(expand=True)[0]
df['msg_send']
df.head()
```
